```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("KMeansClustering.csv")
data.head()
```

```python
data.describe()
```

```python
# Scatter plotting 'Sales' vs 'Rate of Return'
plt.figure(figsize=(10, 6))
plt.scatter(data['Sales'], data['Rate of Return'],  c=data['Cluster'],marker='o')
plt.colorbar(label='Cluster')
plt.xlabel('Sales')
plt.ylabel('Rate of Return')
plt.title('Scatter plot of Sales vs Rate of Return')
plt.grid(True)
plt.show()
```
