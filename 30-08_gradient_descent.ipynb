```python
import numpy as np
import matplotlib.pyplot as plt

X=np.arange(10)
Y=(X-5)**2
print(X,Y)
```

```python
#VISUALISATION of above
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("F(x)")
plt.show()
```

```python
x=0
lr=0.1
error=[]
#50 steps in downhill direction
plt.plot(X,Y)
for i in range(50):
    grad=2*(x-5)
    x=x-lr*grad
    y=(x-5)**2
    error.append(y)
    plt.scatter(x,y)
    #plt.plot(x,y)
    print(x)
    
```

```python
#plotting values of error
plt.plot(error)
```
