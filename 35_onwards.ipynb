```python
#ARGS
def add(*args):
    sum=0
    for i in args:
        sum+=i
        return sum
    print(add(1,2,3,4,5))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y=6+3* np.sin(2*x)

#plot
fig, a=plt.subplots()

a.plot(x,y,linewidth=1.0)

a.set(xlim=(0,8), xticks.np.arrange(1,8),
      ylim=(0,8), yticks.np.arrange(1,8))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
Temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
Res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])

plt.title("Thermistor characteristics")
plt.xlabel("Temp")
plt.ylabel("RES")
plt.plot(Temp,Res)
plt.grid()
#plt.plot(Temp,Res,linestyle='dotted')
plt.show()
```
